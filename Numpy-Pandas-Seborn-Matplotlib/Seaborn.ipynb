# Scatter Plot | Part 3 | Complete Seaborn Tutorial | Beginner to Advanced


### Scatter plot

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
sns.scatterplot(x='age', y='fare', data=titanic_df)
```

```python
sns.scatterplot(x='age', y='fare', data=titanic_df, hue='alive')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age', y='fare', data=titanic_df, hue='alive', style='class')
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age', y='fare', data=titanic_df, hue='alive', style='class', palette = 'inferno')
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age', y='fare', data=titanic_df, hue='alive', style='class',palette='gist_rainbow', alpha=0.5)
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age', y='fare', data=titanic_df, hue='alive', style='class')
sns.lineplot(x='age',y='fare',data=titanic_df,color='green')
plt.title('Titanic Data Analysis')
```

# Bar Plot | Part 4 | Complete Seaborn Tutorial | Beginner to Advanced


### Bar Plots

```python
titanic_df.head(1)
```

```python
sns.barplot(x='class', y='fare', data =titanic_df)
```

A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.

```python
sns.barplot(x='class', y='fare', data=titanic_df, hue='sex')
```

```python
sns.barplot(x='class',y='fare',data=titanic_df,hue='sex',palette='icefire')
```

```python
#orient
sns.barplot(y='class',x='fare',data=titanic_df,hue='sex',palette='inferno',orient='h')
```

```python
#estimator
sns.barplot(x='class',y='fare',data=titanic_df,hue='sex',palette='inferno',estimator=np.max)
```

```python
#estimator
sns.barplot(x='class',y='fare',data=titanic_df,hue='sex',palette='inferno',estimator=np.mean)
```

```python
#confidence interval
sns.barplot(x='class',y='fare',data=titanic_df,hue='sex',palette='inferno',ci=100,errcolor='red',errwidth=10)
```

```python
#saturation
sns.barplot(x='class',y='fare',data=titanic_df,hue='sex',palette='inferno',saturation=0.9)
```

# Heatmaps in Seaborn | Part 5 | Complete Seaborn Series


### Heatmaps

```python
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
# flight_df=flight_df.pivot("month","year","passengers")
flight_df = flight_df.pivot(index="month", columns="year", values="passengers")
flight_df.head()
```

```python
plt.figure(figsize=(12,6))
ax=sns.heatmap(flight_df)
ax
```

```python
plt.figure(figsize=(14,8))
ax=sns.heatmap(flight_df,annot=True,fmt='d')
ax
```

```python
plt.figure(figsize=(14,8))
ax=sns.heatmap(flight_df,annot=True,fmt='d', linecolor='k', linewidths='5')
ax
```

```python
plt.figure(figsize=(14,8))
ax=sns.heatmap(flight_df,annot=True,fmt='d', linecolor='k', linewidths='5',cmap='Reds')
ax
```

```python
plt.figure(figsize=(14,8))
ax=sns.heatmap(flight_df,annot=True,fmt='d', linecolor='k', linewidths='5',cmap='Blues')
ax
```

```python
plt.figure(figsize=(14,8))
ax=sns.heatmap(flight_df,cbar=False)
ax
```

```python
grid_kws = {"height_ratios":(.4,.05),"hspace":.4}
f, (ax,cbar_ax) = plt.subplots(2,gridspec_kw=grid_kws)
ax = sns.heatmap(flight_df, cbar_kws={"orientation":"horizontal"}, ax=ax,
                cbar_ax=cbar_ax,)
```

# Pair Plots in Seaborn | Part 6 | Data Visualisation | Seaborn Complete Playlist


### Pair Plot

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins)
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins, hue='sex', palette='Reds')
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins, hue='species', palette='Blues')
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins, kind='kde', palette='Greens')
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins, kind='hist', palette='Greens')
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,hue='species',diag_kind='hist', palette='rainbow')
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,corner=True)
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,hue='species', markers=['o','s','D'], palette = 'inferno')
```
